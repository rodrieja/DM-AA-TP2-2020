## Funciones

### calculate_feature

```python
def calculate_features(filename, n_mfcc=12,delta=True,deltadelta=True,energy=True, summary_fn = [np.mean, np.std], summary_names=['mean','std']):
  #Abro el archivo:
  x, sr = librosa.core.load(filename,sr=None)
  
  #Calculo MFCCs
  features = librosa.feature.mfcc(x,sr=sr,n_mfcc=n_mfcc)
  feat_names = ['mfcc_{}'.format(i) for i in range(n_mfcc)]
  #Calculo energia:
  if energy:
    energy = librosa.feature.rmse(x)
    features = np.concatenate([features,energy])
    feat_names = feat_names + ['energy']
  #Aplico media y desvio estandar por defecto
  summary_features = np.concatenate([fn(features,axis=1) for fn in summary_fn])
  feat_names = ['{}_{}'.format(name_i,summ_i) for summ_i in summary_names for name_i in feat_names]
  
  #Lo mismo con los delta
  if delta:
    deltafeatures = np.diff(features)
    summary_features = np.concatenate([summary_features,np.concatenate([fn(deltafeatures,axis=1) for fn in summary_fn])])
    d_names = ['d{}'.format(name) for name in feat_names]
  else:
    d_names = []

  #Y con los delta de segundo orden
  if deltadelta:
    deltadeltafeatures = np.diff(features,n=2)
    summary_features = np.concatenate([summary_features,np.concatenate([fn(deltadeltafeatures,axis=1) for fn in summary_fn])]) 
    dd_names = ['dd{}'.format(name) for name in feat_names]
  else:
    dd_names = []

  feat_names = feat_names + d_names + dd_names
  
  return summary_features, feat_names  
```

### HeatmapMatrix

```python
def HeatmapMatrix(title, predictionValues, realValues):
  fig, ((ax, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(25, 13))

  sns.heatmap(confusion_matrix(realValues[0], predictionValues[0], 
                              labels=realValues[0].unique()), annot=True, fmt='g', ax=ax)

  sns.heatmap(confusion_matrix(realValues[1], predictionValues[1], 
                              labels=realValues[1].unique()), annot=True, fmt='g', ax=ax2)

  sns.heatmap(confusion_matrix(realValues[2], predictionValues[2], 
                              labels=realValues[2].unique()), annot=True, fmt='g', ax=ax3)

  sns.heatmap(confusion_matrix(realValues[3], predictionValues[3], 
                              labels=realValues[3].unique()), annot=True, fmt='g', ax=ax4)

  sns.heatmap(confusion_matrix(realValues[4], predictionValues[4], 
                              labels=realValues[4].unique()), annot=True, fmt='g', ax=ax5)

  sns.heatmap(confusion_matrix(realValues[5], predictionValues[5], 
                              labels=realValues[5].unique()), annot=True, fmt='g', ax=ax6)

  title_size = 20
  axes_size = 16
  fig.set_facecolor(color='white')
  fig.suptitle(title, fontsize=24, fontweight='medium')

  ax.set_title('Trainning', fontdict={'fontsize': title_size, 'fontweight': 'medium'})
  ax.set_ylabel('Real', fontdict={'fontsize': axes_size})

  ax2.set_title('Test', fontdict={'fontsize': title_size, 'fontweight': 'medium'})
  ax3.set_title('Validation', fontdict={'fontsize': title_size, 'fontweight': 'medium'})
  ax4.set_title('Records', fontdict={'fontsize': title_size, 'fontweight': 'medium'})
  ax4.set_ylabel('Real', fontdict={'fontsize': axes_size})
  ax4.set_xlabel('Predicted', fontdict={'fontsize': axes_size})
  ax5.set_title('Gaussian Noise', fontdict={'fontsize': title_size, 'fontweight': 'medium'})
  ax5.set_xlabel('Predicted', fontdict={'fontsize': axes_size})
  ax6.set_title('Environment Noise', fontdict={'fontsize': title_size, 'fontweight': 'medium'})
  ax6.set_xlabel('Predicted', fontdict={'fontsize': axes_size})

  plt.show()
```

### CountplotSummarized

```python
def CountplotSummarized(title, predictionValues):
  traLabel = ['Train']*len(predictionValues[0])
  tesLabel = ['Test']*len(predictionValues[1])
  valLabel = ['Validation']*len(predictionValues[2])
  recLabel = ['Records']*len(predictionValues[3])
  gauLabel = ['Gaussian']*len(predictionValues[4])
  envLabel = ['Environment']*len(predictionValues[5])

  dfTra = pd.DataFrame({'subset': traLabel, 'result': predictionValues[0]})
  dfTes = pd.DataFrame({'subset': tesLabel, 'result': predictionValues[1]})
  dfVal = pd.DataFrame({'subset': valLabel, 'result': predictionValues[2]})
  dfRec = pd.DataFrame({'subset': recLabel, 'result': predictionValues[3]})
  dfGau = pd.DataFrame({'subset': gauLabel, 'result': predictionValues[4]})
  dfEnv = pd.DataFrame({'subset': envLabel, 'result': predictionValues[5]})

  DF_Complete = pd.concat([dfTra, dfTes, dfVal, dfRec, dfGau, dfEnv], axis=0)

  title_size = 18
  axes_size = 14

  fig = plt.figure(figsize=(10,8))
  fig.set_facecolor(color='white')

  sns.countplot(data = DF_Complete, x='result', hue='subset')
  plt.xlabel('Numbers', fontdict={'fontsize': axes_size})
  plt.ylabel('Count', fontdict={'fontsize': axes_size})
  plt.title(title, fontdict={'fontsize': title_size, 'fontweight': 'medium'})

  plt.show()
```

## Gradient boosting

Esta es la versión donde selecciono parámetros al azar de una distribución, los uso para entrenar en train y testear en validación ( una vez por conjunto de parámetros). Finalmente, el mejor modelo se evalua en test. Usaria esta version, no la que esta al final

```python
rng = np.random.RandomState(0)
param_grid = {'n_estimators':range(20, 200, 5), 'max_depth':range(1,7), 'learning_rate':np.arange(0,1,0.001)}
# >>> param_grid = {'max_depth':[1, 2], 'b': expon()}
param_list = list(ParameterSampler(param_grid, n_iter=5,
                                    random_state=42))

print(param_list) #
```

```python
## Ojo con esta celda tarda una hora y 20 en correr
##
import time
start = time.time()
estimators_list = []
train_score_list = []
val_score_list = []
print("time in min.:", (time.time() - start)/60)

# ## Para hacer pruebas del loop
# train_f = [[0, 0], [1, 1], [2, 2]]
# test_f = [0, 1, 2]


for i in range(0, len(param_list)):
  ## Establezco un clasificador para cada conjunto de parametros
  clf = GradientBoostingClassifier()
  clf.set_params(**param_list[i])
  clf.fit(X_train, y_train)
  estimators_list.append(clf)

  ## Guardo los valores para elegir el mejor
  train_score_list.append(accuracy_score(y_train, clf.predict(X_train)))
  val_score_list.append(accuracy_score(y_val, clf.predict(X_val)))

print(train_score_list)
print(val_score_list)

print("time in min.:", (time.time() - start)/60)
```

```python
best_index = val_score_list.index(max(val_score_list)) ## de el mer score sobre el set de validación
mejor_GB = estimators_list[best_index]
```

Definimos una nueva función calculate features que agregue ruido a los audios originales. Se probaron diferentes niveles de ruido y finalmente se realizaron 4 iteraciones ajustando el ponderador de ruido: 0.01, 0.04, 0.07 y 0.1

```python
def calculate_features_ruido(filename, n_mfcc=12,delta=True,deltadelta=True,energy=True, summary_fn = [np.mean, np.std], summary_names=['mean','std']):
  #Abro el archivo:
  x, sr = librosa.core.load(filename,sr=None)
  ruido = np.random.normal(size=x.shape)
  x = 0.01*ruido+x
  
  #Calculo MFCCs
  features = librosa.feature.mfcc(x,sr=sr,n_mfcc=n_mfcc)
  feat_names = ['mfcc_{}'.format(i) for i in range(n_mfcc)]
  #Calculo energia:
  if energy:
    energy = librosa.feature.rmse(x)
    features = np.concatenate([features,energy])
    feat_names = feat_names + ['energy']
  #Aplico media y desvio estandar por defecto
  summary_features = np.concatenate([fn(features,axis=1) for fn in summary_fn])
  feat_names = ['{}_{}'.format(name_i,summ_i) for summ_i in summary_names for name_i in feat_names]
  
  #Lo mismo con los delta
  if delta:
    deltafeatures = np.diff(features)
    summary_features = np.concatenate([summary_features,np.concatenate([fn(deltafeatures,axis=1) for fn in summary_fn])])
    d_names = ['d{}'.format(name) for name in feat_names]
  else:
    d_names = []

  #Y con los delta de segundo orden
  if deltadelta:
    deltadeltafeatures = np.diff(features,n=2)
    summary_features = np.concatenate([summary_features,np.concatenate([fn(deltadeltafeatures,axis=1) for fn in summary_fn])]) 
    dd_names = ['dd{}'.format(name) for name in feat_names]
  else:
    dd_names = []

  feat_names = feat_names + d_names + dd_names
  
  return summary_features, feat_names
```
